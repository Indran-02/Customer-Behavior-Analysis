```python
import pandas as pd
df = pd.read_excel('customer_shopping_behavior.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.isnull().sum()
```

```python
df['review_rating'] = df.groupby('category')['review_rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
labels = ['Young Adults', 'Adults', 'Middle-Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
df[['age', 'age_group']].head()
```

```python
frequency_purchase = { 'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90}
df['frequency_purchase_days'] = df['frequency_of_purchases'].map(frequency_purchase)
df[['frequency_of_purchases', 'frequency_purchase_days']].head()
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
df.drop('promo_code_used', axis=1)
```
